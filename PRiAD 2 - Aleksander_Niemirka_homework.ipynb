> **Zadanie 2** Wyświetl wykres słupkowy pokazujący liczbę ludności wszystkich dużych państw świata tj. liczących więcej niż 35 milionów mieszkańców w kolejności od najmniejszego do największego. 

```python
# rozwiązanie zadania
plt.rcParams['figure.dpi'] = 140
panstwa = pd.read_excel('alphabetical-list-of-countries2020.xlsx',header = 4,usecols = range(2,5), index_col = 0)
panstwa.columns = ['Ludnosc','Powierzchnia']
panstwa.index.name = 'Kraj'
panstwa = panstwa.sort_values(by ='Ludnosc')
panstwa = panstwa[panstwa.Ludnosc > 35 * 10**6]
plt.rcParams.update({'font.size': 5.5})
panstwa.plot(kind = 'barh')
```

> **Zadanie 3** Narysuj histogramy poszczególnych atrybutów zbioru danych `iris` z podziałem na klasy, oraz macierz wykresów punktowych. Oceń przydatność poszczególnych atrybutów do rozróżniania odmian irysa

```python
# rozwiązanie zadania
plt.figure(figsize=(10,10), dpi = 120)

df = pd.read_csv("iris.csv")

def pokaz_jeden_atrybut(atr):
    sns.distplot(df.loc[df['species'] == 'setosa', atr], label = 'setosa')
    sns.distplot(df.loc[df['species'] == 'versicolor', atr], label = 'versicolor')
    sns.distplot(df.loc[df['species'] == 'virginica', atr], label = 'viriginica')
    

    
#jest masa warningow ale w skryptach do nauki tez jest ich sporo a wiec zakladam ze jest okej
plt.subplot(2,2,1)
atr1 = 'sepal_length'
pokaz_jeden_atrybut(atr1)

plt.subplot(2,2,2)
atr2 = 'sepal_width'
pokaz_jeden_atrybut(atr2)

plt.subplot(2,2,3)
atr3 = 'petal_length'
pokaz_jeden_atrybut(atr3)

plt.subplot(2,2,4)
atr4 = 'petal_width'
pokaz_jeden_atrybut(atr4)


sns.pairplot(df, kind="scatter", hue = "species")
plt.show()

#okolo 8 godzin pracy :)
```
